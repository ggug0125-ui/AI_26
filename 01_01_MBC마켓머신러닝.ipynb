```python
# 머신러닝 알고리즘은 지도학습,비지도학습으로 나누어짐
# 지도학습: 훈련하기 위한 데이터와 정답이 필요함
# 지도학습에서 데이터와 정답을 입력 (input)과 타킷(target)이라고 함
# 이 둘을 합쳐서 훈련데이터(tarining data)라고 함
# 입력으로 사용된 길이와 무게 -> 특성 (feature)

# MBC마켓에서 살아있는 생선을 최초로 판매 하려고 한다.
# 고객이 온라인 주문을 하면 가장 빠른 물류 센타에서 신선한 생선을 곧바로 배송
# 여기서 문제 발생 :  생선 포장 직원이 생선의 종류를 알지 못함!!
# IP카메라를 이용해서 수족관을 실시간 촬영을 하고 이것을 객체 추적 기술을 사용

# 우선 도미와 빙어 2개의 객체를 찾도록 훈련시작!!!!!!
# 1. 전처리 데이터가 필요함 : 도미 생선의 길이, 무게를 학습을 시켜보자.

```

```python
# 도미 생선의 길이 : bream_length 도미용 길이 리스트
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미 생선의 무게 : bream_weigh 도미용 무게 리스트
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# 특성(feature): 데이터의 특징
# 첫번째 도미의 특성 (길이 : 25.4cm, 무게 : 242.0g )
# 두번째 도미의 특성 (길이 : 26.3cm, 무게 : 290.0g )

# 두 특성을 숫자로 보는것보다 그래프로 표현하면 데이터를 잘 이해 할 수 있다.
# 길이 :x축, 무게 :y축 -> 산점도 (두 좌표의 관계 )
```

```python
import matplotlib.pyplot as plt
# 파이썬에서 과학 계산용 그래프를 그리는 대표적인 패키지 : matplotlib
# https://wikidocs.net/124976 matplotlib 사용법

plt.scatter (bream_length, bream_weight) # 차트에 값을 넣음
# scatter 차트의 종류 (X축 : bream_length, y축 : bream_weight)
plt.xlabel('length') # x축의 이름표
plt.ylabel('weigth') # y축의 이름표
plt.show()
# 도미 35마리를 2차원 그래프에 점으로 나타내봄
# 생선의 길이가 길수록 무게가 많이 나간다~!!! -> 선형적이라고함
```

```python
# 빙어 데이터 14개 준비

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(smelt_length, smelt_weight) # 차트에 값을 넣음
plt.xlabel('length') # x축의 이름표
plt.ylabel('weigth') # y축의 이름표
plt.show()
```

```python
# 이 2개의 리스트를 합쳐서 ai 학습용 자료로 만들어야함

length = bream_length + smelt_length # 도미길이리스트 + 빙어길이 리스트
weight = bream_weight + smelt_weight # 도미무게리스트 + 빙어무게리스트

# 사이킷런을 이용한 ai트레이닝 시작~!!
# 사이킷런의 입력은 2차원 리스트로 만들어야함
print(length)
print(weight)
```

```python

# zip() 나열된 리스트 각각에서 하나씩 원소를 꺼내어 반환
fish_date = [[l,w] for l,w in zip(length, weight)]
# length, weight를 2차원 리스트로 만듬
print(fish_date) # 2차원리스트 출력 해봄
```

```python
# 머신러닝은 정답을 target 이라는 값으로 알려줘야한다
# target 0 (빙어), 1 (도미)을 처리하려함

# 생선 49개에 대한 정답 리스트를 만들어야함
# 머신러닝에서는 정답데이터를 2진분류로 테스트 해볼까함 (1: 도미, 0:빙어)

fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# pip install scikit-learn 파이참에서는 설치 후 구동해야함!
# 전처리 데이터가 완성 되었으니 사이킷런 패키지를 활용해서 트레이닝을 해보자

from sklearn.neighbors import KNeighborsClassifier  #사이킷런 패키지안에. 이웃패키지
# 사이킷런 패키지.이웃모듈    KNeighborsClassifier  k-최근접 이웃 알고리즘
#      대문자로 시작하면 (클래스)  최대5개  클래스피어는 이진분류용

kn = KNeighborsClassifier() # 임포트된 객체를 생성하여 변수에 연결
# kn. 하면 k-최근접 알고리즘을 사용한다.
# kn.fit()훈련시작
# kn.score() 평가->점수보기
# kn.predict 결과 -> 샘플 넣어서 2진 비교
```

```python
# fish_data, fish_target을 전달하여 도미를 찾기 위한 기준을 학습(.fit) 시킨다.
# (트레이닝데이터 : 훈련정답데이터)

kn.fit(fish_date, fish_target) # 훈련 끝
```

```python
# kn 얼마나 잘 훈련되었는지 평가를 해보자 : score()
kn.score(fish_date, fish_target)  # 1,0 - > 100%(정확도)
# 현재 더미데이터가 많지 않아 훈련용과 평가용 데이터가 같음으로 100%의 결과가 나옴
# 차후에는 더미데이터를 쪼개서 평가하는것이 좋다.
```

```python
# k-최근접 이웃 알고리즘 파악해보기
# 어떤 데이터에 대한 답을 구할때 주위의 다른 데이터를 보고
# 다수를 차지하는 것을 정답으로 출력
```

화면 캡처 2026-02-23 114546.png

```python
kn.predict([[30,600]])  #.predict[길이,무게] 새로운데이터의 정답을 예측(2진분류)
# 결론 1 - > 도미
```

```python
kn.predict([[8,10]]) # .predict[길이,무게] 새로운 데이터릐 정답을 예측(2진분류)
# 결론 0 - > 빙어
```

```python
# K-최근접 이웃 알고리즘을 위해 준비해야 할 일은 데이터를 모두 가지고 있는게 전부~!!
# 새로운 데이터에 대해 예측할 때는 가장 가까운 직선거리에 어떤 데이터가 있는지 살피기만 함
# 단점, 데이터가 아주 많은 경우 사용하기 좀 어렵다
# (데이터가 크기 때문에 메모리가 많이 필요, 직선거리 계산 시간도 오래 걸림)
# 객체를 메모리에 만들고 활용한다. -> 재부팅시 fit를 다시해야함 (안그럴려면 파일로 저장!) 메모리는 종료시 다지워진다~ 새로시작시 다시실행후 이용해
print(kn._fit_X) # _fit_X 내장된 변후로 훈련데이터를 보관함!
```

```python
print(kn._y) # 정답파일 1:도미, 0 : 빙어
```

```python
# 결론 k-최근접 이웃 알고리즘은 훈련 결과파일이 존재하지 않는다.(메모리에 로드 ->재부팅시 사라짐)
# 새로운 데이터가 들어오면 가장 가까운데이터를 메모리에서 참고하여 도미, 빙어를 구분함
kn.predict([[5,15]]) # 길이가 5cm,무게 15g인 값을 판단하면 0:빙어
```

```python
print(kn._fit_method)
```

```python
# 객체 생성시 n_neighbore=??? 참고할 이웃 데이터 수 기본 값은 5
# 이웃 객체를 5개 찾아서 결론을 낸다.

# 만약 5개가 아닌 다른값으로 변경하여 데이터를 훈련하고 싶으면???

kn49 =  KNeighborsClassifier(n_neighbors=49) #참고 데이터를 49로 한 모델 kn49변수
#       KNeighborsClassifier(n_neighbors=5) 기본값 - > KNeighborsClassifier()

# 원본타겟, 원본데이터가 49인데 가장 가까운데이터 49개를 사용함 (전체 데이터가 이웃??)

kn49.fit(fish_date, fish_target)
kn49.score(fish_date, fish_target)  # 훈련결과 71%로 좋지 않는 성능을 보임

```

```python
# n_neighbore = 값을 변경해 가면서 최적의 정답을 찾는것이 머신러닝 훈련개발자의 목료
kn49.predict([[5,15]]) # 길이 5cm, 무게 15g인 데이터를 찾아보자. 49개를 참고한 데이터에서 값을 찾는거야`!!!!
# 원래는 빙어값으로 나와야하는데 값은 도미로 나온다~!!!
```
